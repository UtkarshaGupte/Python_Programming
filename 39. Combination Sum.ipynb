```python
'''
39. Combination Sum

Given an array of distinct integers candidates and a target integer target, 
return a list of all unique combinations of candidates where the chosen numbers sum to target. 
You may return the combinations in any order.

The same number may be chosen from candidates an unlimited number of times. 
Two combinations are unique if the frequency of at least one of the chosen numbers is different.

It is guaranteed that the number of unique combinations that sum up to target is less than 150 combinations 
for the given input.
 
Example 1:
Input: candidates = [2,3,6,7], target = 7
Output: [[2,2,3],[7]]
Explanation:
2 and 3 are candidates, and 2 + 2 + 3 = 7. Note that 2 can be used multiple times.
7 is a candidate, and 7 = 7.
These are the only two combinations.

Example 2:
Input: candidates = [2,3,5], target = 8
Output: [[2,2,2,2],[2,3,3],[3,5]]

Example 3:
Input: candidates = [2], target = 1
Output: []

Note:

Why a deep copy is needed?

Without a deep copy, we are appending a reference to a list that is constantly being appended and 
ultimately all the elements are being popped out from it hence making a deep copy is required.
'''

class Solution:
    def combinationSum(self, candidates: List[int], target: int) -> List[List[int]]:
        
        res = []
        
        def dfs(currComb, remainSum, start):
            
            if remainSum == 0:
                res.append(copy.deepcopy(currComb))
                return
            
            if remainSum < 0:
                return
            
            
            for i in range(start, len(candidates)):
                currComb.append(candidates[i])
                 
                dfs(currComb, remainSum - candidates[i], i)
                
                currComb.pop()
                
                
        dfs([], target, 0)
        
        return res
    
    
```
