```python
'''
2128. Remove All Ones With Row and Column Flips

You are given an m x n binary matrix grid.

In one operation, you can choose any row or column and flip each value in that row or column 
(i.e., changing all 0's to 1's, and all 1's to 0's).

Return true if it is possible to remove all 1's from grid using any number of operations or false otherwise.


Example 1:
Input: grid = [[0,1,0],[1,0,1],[0,1,0]]
Output: true
Explanation: One possible way to remove all 1's from grid is to:
- Flip the middle row
- Flip the middle column

Example 2:
Input: grid = [[1,1,0],[0,0,0],[0,0,0]]
Output: false
Explanation: It is impossible to remove all 1's from grid.

Example 3:
Input: grid = [[0]]
Output: true
Explanation: There are no 1's in grid.

Time/Space Complexity:

Time - O(rows*cols) +O(rows)

Space - O(1)


'''

class Solution:
    def removeOnes(self, grid: List[List[int]]) -> bool:
                
        rows, cols = len(grid), len(grid[0])
        
        
        for c in range(cols):
            
            temp = grid[0][c]
            
            if temp == 1:
                
                for r in range(rows):
                    
                    grid[r][c] = abs(grid[r][c] - 1)
                           
                    
        for r in range(rows):
            
            print(sum(grid[r]))
            
            
            if sum(grid[r]) == 0 or sum(grid[r]) == cols:
                continue
                
            else:
                
                return False
        
        return True
    
    
```
