```python
'''
148. Sort List
Given the head of a linked list, return the list after sorting it in ascending order.

Example 1:
Input: head = [4,2,1,3]
Output: [1,2,3,4]

Example 2:
Input: head = [-1,5,3,4,0]
Output: [-1,0,3,4,5]

Example 3:
Input: head = []
Output: []

Time/Space Complexity:

Time - O(logN) - to get single nodes, O(logN) to merge it, sorting will take place in O(N) hence O(NlogN)

Space - O(1)
'''

class Solution:
    def sortList(self, head: Optional[ListNode]) -> Optional[ListNode]:
        
        if head is None or head.next is None:
            return head
        
        l1 = head
        slow, fast = head, head.next
        
        while fast and fast.next:           
            slow = slow.next
            fast = fast.next.next
            
        mid = slow
        l2 = mid.next
        mid.next = None
        
        Node1 = self.sortList(l1)
        Node2 = self.sortList(l2)
        
        return self.mergeList(Node1, Node2)        
        
    def mergeList(self, headOne, headTwo):
        
        temp = ListNode(0)
        res = temp
        while headOne and headTwo:
            if headOne.val <= headTwo.val:
                temp.next = headOne
                headOne = headOne.next
            else:
                temp.next = headTwo
                headTwo = headTwo.next
                    
            temp = temp.next
        
        if headOne:
            temp.next = headOne
        elif headTwo:
            temp.next = headTwo
            
        return res.next      
    
    
    
     
```
