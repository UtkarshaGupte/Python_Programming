```python
'''
1065. Index Pairs of a String
Given a string text and an array of strings words, return an array of all index pairs [i, j] 
so that the substring text[i...j] is in words.

Return the pairs [i, j] in sorted order (i.e., sort them by their first coordinate, 
and in case of ties sort them by their second coordinate).


Example 1:
Input: text = "thestoryofleetcodeandme", words = ["story","fleet","leetcode"]
Output: [[3,7],[9,13],[10,17]]

Example 2:
Input: text = "ababa", words = ["aba","ab"]
Output: [[0,1],[0,2],[2,3],[2,4]]
Explanation: Notice that matches can overlap, see "aba" is found in [0,2] and [2,4].
'''

class TrieNode:
    def __init__(self):
        self.children = {}
        
    def insert(self,word):
        curr = self.children
        
        for char in word:            
            if char not in curr:
                curr[char] = {}
                
            curr = curr[char]
            
        curr['*'] = word
                

class Solution:
    def indexPairs(self, text: str, words: List[str]) -> List[List[int]]:
        trie = TrieNode()
        
        for word in words:
            trie.insert(word)
        
        n = len(text)
        res = []
        
        for i in range(n):
            currIndex = i
            currChild = trie.children
            
            while currIndex < n and text[currIndex] in currChild:
                currChild = currChild[text[currIndex]]
                if '*' in currChild:
                    res.append([i,currIndex])                  
                    
                currIndex += 1
                
        return res
                    
```
