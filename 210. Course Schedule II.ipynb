```python
'''
210. Course Schedule II
There are a total of numCourses courses you have to take, labeled from 0 to numCourses - 1. 
You are given an array prerequisites where prerequisites[i] = [ai, bi] indicates that you must take course bi 
first if you want to take course ai.

For example, the pair [0, 1], indicates that to take course 0 you have to first take course 1.
Return the ordering of courses you should take to finish all courses. If there are many valid answers, 
return any of them. If it is impossible to finish all courses, return an empty array.


Example 1:
Input: numCourses = 2, prerequisites = [[1,0]]
Output: [0,1]
Explanation: There are a total of 2 courses to take. To take course 1 you should have finished course 0. 
So the correct course order is [0,1].

Example 2:
Input: numCourses = 4, prerequisites = [[1,0],[2,0],[3,1],[3,2]]
Output: [0,2,1,3]
Explanation: There are a total of 4 courses to take. To take course 3 you should have finished both courses 1 and 2. 
Both courses 1 and 2 should be taken after you finished course 0.
So one correct course order is [0,1,2,3]. Another correct ordering is [0,2,1,3].

Example 3:
Input: numCourses = 1, prerequisites = []
Output: [0]

Time/Space Complexity:

Time - O(N+E) - The queue contains all zero in-degree nodes and inside the while loop we iterate over all its neighbors.

Space - O(N+E) - In the worst-case scenario, there might be no prereq at all, then all N nodes will have zero in-degrees so O(N). The dictionary will store all edge information hence O(E)



Kahn's Algorithm explanation - https://www.youtube.com/watch?v=cIBFEhD77b4
'''

class Solution:
    def findOrder(self, numCourses: int, prerequisites: List[List[int]]) -> List[int]:
        
        prereq_dict = {i:[] for i in range(numCourses)}
        incoming_edges = {}
        
        for a,b in prerequisites:
            prereq_dict[b].append(a)
            
            incoming_edges[a] = incoming_edges.get(a,0)+1
            
        
        q = deque([])
        
        for node in range(numCourses):
            if node not in incoming_edges:
                q.append(node)
        
        res = []
        
        while q:
            
            currNode = q.popleft()
            res.append(currNode)
            
            for neighbour in prereq_dict[currNode]:
                incoming_edges[neighbour] -= 1
                
                if incoming_edges[neighbour]  == 0:
                    q.append(neighbour)
                
                
        return res if len(res) == numCourses else []
 
```
