```python
'''
23. Merge k Sorted Lists

You are given an array of k linked-lists lists, each linked-list is sorted in ascending order.

Merge all the linked-lists into one sorted linked-list and return it.


Example 1:
Input: lists = [[1,4,5],[1,3,4],[2,6]]
Output: [1,1,2,3,4,4,5,6]
Explanation: The linked-lists are:
[
  1->4->5,
  1->3->4,
  2->6
]
merging them into one sorted list:
1->1->2->3->4->4->5->6

Example 2:
Input: lists = []
Output: []

Example 3:
Input: lists = [[]]
Output: []

Time/Space Complexity:

Time - O(NlogK)- The function mergeTwoLists will take O(N) time where N is the total number of nodes in two lists. 
Since we are combining two lists and merging it, the process will iterate logK times. 
Hence total time complexity will be O(NlogK)

Space - O(N) - To save the temporary merged list.

'''

# Definition for singly-linked list.
# class ListNode:
#     def __init__(self, val=0, next=None):
#         self.val = val
#         self.next = next
class Solution:
    def mergeKLists(self, lists: List[Optional[ListNode]]) -> Optional[ListNode]:        
        
        if len(lists) == 0:
            return None
                
        while len(lists) > 1:
            
            mergedList = []

            for i in range(0,len(lists),2):

                l1 = lists[i]
                l2 = lists[i+1] if (i+1) < len(lists) else None

                mergedList.append(self.mergeTwoLists(l1,l2))

            lists = mergedList
            
        return lists[0]
                           
    def mergeTwoLists(self,l1,l2):
        
        temp = ListNode(0)
        prev = temp
        
        while l1 and l2:
            
            if l1.val <= l2.val:
                prev.next = l1
                l1 = l1.next                
            else:
                prev.next = l2
                l2 = l2.next
                
            prev = prev.next
            
        if l1:
            prev.next = l1
        elif l2:
            prev.next = l2
                     
        return temp.next
    
```
