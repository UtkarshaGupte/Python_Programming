```python
'''
51. N-Queens
The n-queens puzzle is the problem of placing n queens on an n x n chessboard such that no two queens attack each other.

Given an integer n, return all distinct solutions to the n-queens puzzle. You may return the answer in any order.

Each solution contains a distinct board configuration of the n-queens' placement, where 'Q' and '.' 
both indicate a queen and an empty space, respectively.

Example 1:
Input: n = 4
Output: [[".Q..","...Q","Q...","..Q."],["..Q.","Q...","...Q",".Q.."]]
Explanation: There exist two distinct solutions to the 4-queens puzzle as shown above

Example 2:
Input: n = 1
Output: [["Q"]]

Time/Space Complexities:

Time: O(N!)

The first queen can be placed in any of the N boxes, the next queen can be placed in any box except for
the same column as the first and in the positive/negative diagonal of the first.
Hence can be placed in any of the N-2 boxes. Similarly next can be placed in any of the N-4 boxes.

Space: O(N^2)
'''

class Solution:
    def solveNQueens(self, n: int) -> List[List[str]]:
        
        col = set()
        negativediagonal = set()
        positivediagonal = set()
        
        res = []        
        #Dummy for 1 placement solution
        
        board = [["."]*n for row in range(n)]
        
        def backtrack(r):
            
            if r == n:
                solution = ["".join(row) for row in board]
                res.append(solution)
                return
            
            for c in range(n):
                
                if ((r+c) in positivediagonal or #positive diagonals
                    (r-c) in negativediagonal or #negaive diagonals
                     c in col):     #same column 
                    continue
                    
                col.add(c)
                negativediagonal.add(r-c)
                positivediagonal.add(r+c)
                
                board[r][c] = "Q"
                
                backtrack(r+1)
                
                col.remove(c)
                negativediagonal.remove(r-c)
                positivediagonal.remove(r+c)
                board[r][c] = "."
                                       
        backtrack(0)
        
        return res
                
   
```
