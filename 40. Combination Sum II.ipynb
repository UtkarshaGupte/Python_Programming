```python
'''
40. Combination Sum II

Given a collection of candidate numbers (candidates) and a target number (target), 
find all unique combinations in candidates where the candidate numbers sum to target.

Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations. 

Example 1:
Input: candidates = [10,1,2,7,6,1,5], target = 8
Output: 
[
[1,1,6],
[1,2,5],
[1,7],
[2,6]
]

Example 2:
Input: candidates = [2,5,2,1,2], target = 5
Output: 
[
[1,2,2],
[5]
]

#366 days of Daily Coding



Day 281:

Combination Sum II



Problem Statement:

Given a collection of candidate numbers (candidates) and a target number (target), find all unique combinations in candidates where the candidate numbers sum to target.

Each number in candidates may only be used once in the combination.

Note: The solution set must not contain duplicate combinations.

 

Example 1:

Input: candidates = [10,1,2,7,6,1,5], target = 8

Output: 

[

[1,1,6],

[1,2,5],

[1,7],

[2,6]

]



Example 2:

Input: candidates = [2,5,2,1,2], target = 5

Output: 

[

[1,2,2],

[5]

]



Time/Space Complexity:

Time - O(N log N) + O(2^n) - O(N log N) to sort the array. Each element of the array can either be included or
not included in the combination hence 2^n

Space - O(n) - For the recursion stack and storing currCand
'''

class Solution:
    def combinationSum2(self, candidates: List[int], target: int) -> List[List[int]]:
        
        res = []
        
        candidates.sort()
        
        def dfs(currCand, remainSum, start):
            
            if remainSum == 0:
                res.append(currCand[:])
                
            if remainSum < 0:
                return
            
            prev = -1
            for i in range(start, len(candidates)):
                
                if candidates[i] == prev:
                    continue
                
                currCand.append(candidates[i])
                
                dfs(currCand, remainSum - candidates[i], i+1)
                
                currCand.pop()
                
                prev = candidates[i]
                
        dfs([], target, 0)
        
        return res
        
 
```
