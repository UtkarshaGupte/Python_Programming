```python
'''
366. Find Leaves of Binary Tree
Given the root of a binary tree, collect a tree's nodes as if you were doing this:

Collect all the leaf nodes.
Remove all the leaf nodes.
Repeat until the tree is empty.
 

Example 1:
Input: root = [1,2,3,4,5]
Output: [[4,5,3],[2],[1]]
Explanation:
[[3,5,4],[2],[1]] and [[3,4,5],[2],[1]] are also considered correct answers since per each level 
it does not matter the order on which elements are returned.

Example 2:
Input: root = [1]
Output: [[1]]

Time/Space Complexity:

Time - O(N)

Space - O(N)
 
'''

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def findLeaves(self, root: Optional[TreeNode]) -> List[List[int]]:
        
        
        height_dict = defaultdict(list)
        
        
        def dfs(currNode):
            
            
            if currNode.left is None and currNode.right is None:
                height_dict[0].append(currNode.val)
                return 0
            
            left_currHeight, right_currHeight = 0 , 0
                
            if currNode.left:
                left_currHeight = dfs(currNode.left)
            
            
            if currNode.right:
                right_currHeight = dfs(currNode.right)
                
            
            currHeight = 1 + max(left_currHeight, right_currHeight)
            height_dict[currHeight].append(currNode.val)  
            
            
            return currHeight
        
        
        
        dfs(root)
        
        return height_dict.values()
                
        
```
