```python
'''
227. Basic Calculator II
Given a string s which represents an expression, evaluate this expression and return its value. 

The integer division should truncate toward zero.

You may assume that the given expression is always valid. 
All intermediate results will be in the range of [-231, 231 - 1].

Note: You are not allowed to use any built-in function which evaluates strings as mathematical expressions, such as eval().


Example 1:
Input: s = "3+2*2"
Output: 7

Example 2:
Input: s = " 3/2 "
Output: 1

Time/Space Complexities:

Time: O(N)

Space: O(N)
'''

class Solution:
    def calculate(self, s: str) -> int:
        s = s.strip()
        s_len = len(s)
        
        stack = []
        
        currNumber = ""
        currOperation = "+"
        count = 0
             
        for char in s:
            count += 1
            
            if char.isdigit():                
                currNumber += char
                
            if (char.isdigit() and count == s_len) or (not char.isdigit() and char != " "):
                if currOperation == "-":
                    stack.append(-int(currNumber))
                    
                elif currOperation == "+":
                    stack.append(int(currNumber))
                    
                elif currOperation == "*":
                    prevNumber = stack.pop()
                    res = prevNumber * int(currNumber)
                    stack.append(res)
                
                else:
                    prevNumber = stack.pop()
                    
                    if prevNumber < 0:
                        res = - (abs(prevNumber) // int(currNumber))                       
                    else:
                        res = prevNumber // int(currNumber)
                        
                    stack.append(res)
                    
                currOperation = char
                currNumber = ""                
                
        result = 0
        
        while stack:
            result += stack.pop()
            
        return result
    
```
