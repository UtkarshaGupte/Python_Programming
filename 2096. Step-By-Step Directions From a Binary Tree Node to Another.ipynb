```python
'''
2096. Step-By-Step Directions From a Binary Tree Node to Another

You are given the root of a binary tree with n nodes. Each node is uniquely assigned a value from 1 to n.
You are also given an integer startValue representing the value of the start node s, and a different integer 
destValue representing the value of the destination node t.

Find the shortest path starting from node s and ending at node t. 
Generate step-by-step directions of such path as a string consisting of only the uppercase letters 'L', 'R', and 'U'. 
Each letter indicates a specific direction:

'L' means to go from a node to its left child node.
'R' means to go from a node to its right child node.
'U' means to go from a node to its parent node.
Return the step-by-step directions of the shortest path from node s to node t.


Example 1:
Input: root = [5,1,2,3,null,6,4], startValue = 3, destValue = 6
Output: "UURL"
Explanation: The shortest path is: 3 → 1 → 5 → 2 → 6.

Example 2:
Input: root = [2,1], startValue = 2, destValue = 1
Output: "L"
Explanation: The shortest path is: 2 → 1.

Time/Space Complexity:

Time - O(N)

Space - O(N)
'''

# Definition for a binary tree node.
# class TreeNode:
#     def __init__(self, val=0, left=None, right=None):
#         self.val = val
#         self.left = left
#         self.right = right
class Solution:
    def getDirections(self, root: Optional[TreeNode], startValue: int, destValue: int) -> str:
        
        graph_dict = defaultdict(list)
        
        q = deque([root])
               
        while q:
            currNode = q.popleft()
            
            if currNode.left:
                graph_dict[currNode.val].append((currNode.left.val, 'L'))
                graph_dict[currNode.left.val].append((currNode.val,'U'))
                q.append(currNode.left)
                
            if currNode.right:
                graph_dict[currNode.val].append((currNode.right.val, 'R'))
                graph_dict[currNode.right.val].append((currNode.val,'U'))
                q.append(currNode.right)
        
        print(graph_dict)
        isVisited = set()
        
        #BFS finds the shortest path in undirected graph problems
        q = deque([(startValue, "")])
        
        while q:
            currNode, currPath = q.popleft()
            
            if currNode == destValue:
                return currPath
            if currNode in isVisited:
                continue
                
            isVisited.add(currNode)
            
            for neighbour, path in graph_dict[currNode]:
                if neighbour not in isVisited:
                    q.append((neighbour, currPath+path))
            
            
```
